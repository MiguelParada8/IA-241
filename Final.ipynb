# Import Data

```python
df = pandas.read_excel('s3://paradaia241/COVID-19-geographic-disbtribution-worldwide.xlsx')
df[:10] #the top 10 rows
```

# COVID-19 Explanation: 
COVID-19 is a a respiratory disease that is spreading gloabally, infecting millsions of people. Nation's governments are issuing isolation procedures to stop the spread of virus. National research labs are working quickly to construct a vaccine, which might take up to two years.

# 1. Top 20 Countries with Highest Number of Cases

```python
sum_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
sum_cases_by_country.nlargest(20).plot.bar()
```

# Interpretation:
This graph shows the Top 20 Countries with the most positive cases of COVID-19. 

# 2. Select Spain Data

```python
Spain_data = df.loc[df['countriesAndTerritories'] == 'Spain']
Spain_data[:15] #the top 15 rows
```

# Spain's Death Rate Increase Per Month

```python
Spain_data = df.groupby('dateRep').sum()['deaths']

Spain_data.plot()
```

# Interpretation:
The line graph exhibits the death rate in Spain, showing that April produced the most spikes.  

# 3. Select Turkey Data

```python
Turkey_data = df.loc[df['countriesAndTerritories'] == 'Turkey']
Turkey_data[:15] #the top 15 rows
```

# Turkey's Infection Rate Increase Per Month

```python
Turkey_data=df.groupby('dateRep').mean()['cases']
Turkey_data.plot()
```

# Interpretation:
The line graph exhibits the increase rate in the number of positive COVID-19 cases in Turkey, showing that March produced a steady incline. 

# 4. Number of Deaths Per Continent

```python
global_data = df.groupby('continentExp').sum()['deaths']
global_data.plot.bar()
```

# Interpretation:
The bar graph shows the number COVID-19 related deaths within each continent(continents chosen using the contintentExp Column). Europe is the continent with the most confirmed deaths, as America is the second highest.

# 5. Number of Cases Per Continent

```python
global_data = df.groupby('continentExp').sum()['cases']
global_data.plot.bar()
```

# Interpretation:
The bar graph shows the number of positive COVID-19 cases within each continent(continents chosen using the contintentExp Column). Europe is the continent with the most positive cases, as America is a close second.
